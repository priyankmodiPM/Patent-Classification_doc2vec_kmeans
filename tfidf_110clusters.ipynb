```python
#Import all the dependencies
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize
import sys
import csv
csv.field_size_limit(sys.maxsize)

with open('data1997.csv', 'r') as f:
    reader = csv.reader(f)
    your_list = list(reader)
print(len(your_list))
tagged_data = [TaggedDocument(words=word_tokenize(str(d)), tags=[str(i)]) for i, d in enumerate(your_list)]
max_epochs = 100
vec_size = 128
alpha = 0.025

import gensim.models as g
model="d2v(2).model"
m = g.Doc2Vec.load(model)

from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt

scaler = MinMaxScaler()
X = []
for i in range(5199):
    X.append(m.docvecs[i])
X_scaled = scaler.fit_transform(X)

from sklearn.metrics import accuracy_score #works
from sklearn.metrics import silhouette_score #works
# def km(num_clust):
kmeans =  KMeans(n_clusters=110, max_iter=600, algorithm = 'auto')
kmeans.fit(X_scaled)
KMeans(algorithm='auto', copy_x=True, init='k-means++', max_iter=600,
    n_clusters=110, n_init=10, n_jobs=1, precompute_distances='auto',
    random_state=None, tol=0.0001, verbose=0)
labels = kmeans.predict(X_scaled)    
clusters = {}
clusters_id = {}
n = 0
for item in labels:
    if item in clusters:
        clusters[item].append(your_list[n][2])
    else:
        clusters[item] = [your_list[n][2]]
    n +=1
n = 0
for item in labels:
    if item in clusters_id:
        clusters_id[item].append(your_list[n][1])
    else:
        clusters_id[item] = [your_list[n][1]]
    n +=1
sil_score = silhouette_score(X_scaled, labels, metric='euclidean')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(analyzer='word', ngram_range=(1,5), min_df = 0, stop_words = 'english')
```

```python
tfidf_matrix =  tf.fit_transform(corpus_small)
feature_names = tf.get_feature_names() 
len(feature_names)
```
